```python
import os
import glob

def count_files(folder_path, x_values, z_values):
    total_files = 0
    completed_files = 0

    for x in x_values:
        for z in z_values:
            pattern = os.path.join(folder_path, f"training_history_{x}_50_{z}.csv")
            files = glob.glob(pattern)
            total_files += 1
            if files:
                completed_files += 1

    if total_files == 0:
        return 0.0
    else:
        return (completed_files / total_files) * 100
    
def count_unique_y_values(folder_path, x_values, z_values):
    unique_y_values = set()

    for x in x_values:
        for z in z_values:
            pattern = os.path.join(folder_path, f"training_history_{x}_{'*'}_{z}.csv")
            files = glob.glob(pattern)
            for file in files:
                # Extracting y value from file name
                y_value = file.split('_')[3]  # Corrected index to extract y_value
                unique_y_values.add(y_value)

    return len(unique_y_values)

root_directory = "/home/p/apotoup/DeepNN/LSTM/"
subfolders = ["ANN", "LSTM", "RNN", "GRU"]
x_values = [0, 1, 2, 3]
z_values = ['relu', 'tanh']
max_possible_unique_y_values = 4  # Maximum possible unique y values

print ("Generated Data")
for folder in subfolders:
    folder_path = os.path.join(root_directory, folder)
    if os.path.exists(folder_path):
        completion_percentage = count_files(folder_path, x_values, z_values)
        print(f"Percentage of completion for {folder}: {completion_percentage:.2f}%")

print ("Tuning Percentage")
for folder in subfolders:
    folder_path = os.path.join(root_directory, folder)
    if os.path.exists(folder_path):
        unique_y_count = count_unique_y_values(folder_path, x_values, z_values)
        if folder == 'LSTM' or folder == 'GRU':                     
            percentage = (unique_y_count / 6) * 100
            print(f"Percentage of unique y values for {folder}: {percentage:.2f}%")
            continue
            
        unique_y_count = count_unique_y_values(folder_path, x_values, z_values)
        percentage = (unique_y_count / max_possible_unique_y_values) * 100
        print(f"Percentage of unique y values for {folder}: {percentage:.2f}%")

print ("/---------------------------------/")
print ("Bitcoin")
root_directory = "/home/p/apotoup/DeepNN/LSTM/Bitcoin"
for folder in subfolders:
    folder_path = os.path.join(root_directory, folder)
    if os.path.exists(folder_path):
        if folder == "RNN":
            x_values_temp =[0,1,2,3,4,5]
            completion_percentage = count_files(folder_path, x_values_temp, z_values)
            print(f"Percentage of completion for {folder}: {completion_percentage:.2f}%")
        else:
            completion_percentage = count_files(folder_path, x_values, z_values)
            print(f"Percentage of completion for {folder}: {completion_percentage:.2f}%")
        

print ("Tuning Percentage")
for folder in subfolders:
    folder_path = os.path.join(root_directory, folder)
    if os.path.exists(folder_path):
        unique_y_count = count_unique_y_values(folder_path, x_values, z_values)
        percentage = (unique_y_count / max_possible_unique_y_values) * 100
        print(f"Percentage of unique y values for {folder}: {percentage:.2f}%")

        

root_directory = "/home/p/apotoup/DeepNN/LSTM/Bitcoin/Classification"
completion_percentage = count_files(root_directory, x_values, z_values)
```

```python
import os
import glob

def count_files_cl(folder_path, x_values, y_values, z_values):
    total_files = 0
    completed_files = 0
    found_combinations = set()  # To store found combinations

    for x in x_values:
        for z in z_values:
            for y in y_values:
                pattern = os.path.join(folder_path, f"training_history_{x}_{y}_{z}.csv")
                files = glob.glob(pattern)
                total_files += 1
                if files:
                    completed_files += 1
                    found_combinations.add((x, y, z))  # Add found combination to the set

    if total_files == 0:
        return 0.0
    else:
        completion_percentage = (completed_files / total_files) * 100
        missing_combinations = [(x, y, z) for x in x_values for y in y_values for z in z_values if (x, y, z) not in found_combinations]
        return completion_percentage, missing_combinations

z_values = ['relu', 'tanh']
x_values = [0, 1, 2, 3]
y_values = [50, 500, 1000] 

root_directory = "/home/p/apotoup/DeepNN/LSTM/Bitcoin/Classification"
completion_percentage, missing_combinations = count_files_cl(root_directory, x_values, y_values, z_values)
print(f"Percentage of unique: {completion_percentage:.2f}%")
print("Missing combinations:")
for combination in missing_combinations:
    print(combination)
```
